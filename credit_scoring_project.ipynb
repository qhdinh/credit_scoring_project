```python
import pandas as pd

credits = pd.read_excel("20180301_Sample Data_CreditScoring.xlsx")
```

```python
credits.describe()
```

```python
credits.columns
```

```python
credits = credits.drop('Customer', axis = 1)
```

```python
credits.head()
```

```python
target = credits['DEFAULT']

features = credits.drop('DEFAULT', axis = 1)
```

```python
features.head()
```

```python
from sklearn.preprocessing import StandardScaler

features = StandardScaler().fit_transform(features)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(features)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
principalDf.head()
```

```python
principalDf.plot.scatter(x='principal component 1',
...                       y='principal component 2',
...                       c='DarkBlue')
```

```python
pca.explained_variance_ratio_
```

```python
finalDf = pd.concat([principalDf, target], axis = 1)
```

```python
finalDf.head()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = [0, 1]
colors = ['b', 'r']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['DEFAULT'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
